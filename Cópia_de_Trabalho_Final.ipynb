```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import pickle

# Importar os dados
url_titanic = '/content/drive/MyDrive/titanic.xlsx'
df_titanic = pd.read_excel(url_titanic)

# Tratar os dados
df_titanic.drop(['Name','PassengerId','index'], axis=1, inplace=True)
df_titanic['Sex'] = df_titanic['Sex'].map({'male': 0, 'female': 1})

# Separar variáveis
X = df_titanic[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
y = df_titanic['Survived_y']

x_treino, x_teste, y_treino, y_teste = train_test_split(
    X, y, test_size=0.10, random_state=1
)

# Treinar modelo
modelo = DecisionTreeClassifier(max_depth=1)
modelo.fit(x_treino, y_treino)

# Avaliar modelo
previsoes = modelo.predict(x_teste)
accuracy = accuracy_score(y_teste, previsoes)
print(f'Acurácia: {accuracy}')

# Salvar modelo em formato pickle
with open("modelo_titanic.pkl", "wb") as arquivo:
    pickle.dump(modelo, arquivo)

print("Modelo salvo com sucesso!")

#part 2
with open("modelo_titanic.pkl", "rb") as arquivo:
    modelo_carregado = pickle.load(arquivo)

print("\nDigite as informações do passageiro para prever:")

# Perguntar os dados para o usuário
pclass = int(input("Classe (1 = Primeira, 2 = Segunda, 3 = Terceira): "))
sex = int(input("Sexo (0 = Masculino, 1 = Feminino): "))
age = float(input("Idade: "))
sibsp = int(input("Número de irmãos/cônjuges a bordo: "))
parch = int(input("Número de pais/filhos a bordo: "))
fare = float(input("Valor da passagem: "))

# Criar DataFrame com os dados do usuário
novo_passageiro = pd.DataFrame([[pclass, sex, age, sibsp, parch, fare]],
                               columns=['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare'])

# Fazer previsão
previsao = modelo_carregado.predict(novo_passageiro)

# Mostrar resultado
if previsao[0] == 1:
    print("O passageiro provavelmente sobreviveu.")
else:
    print("O passageiro provavelmente não sobreviveu.")
```
